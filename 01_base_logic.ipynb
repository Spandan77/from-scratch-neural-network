```python
import random
import math
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
def dsigmoid(x):
    s = sigmoid(x)
    return s * (1-s)
```

```python
X = [
    [1, 1, 1, 0, 1, 0, 0, 1, 0], #T
    [1, 1, 1, 0, 1, 0, 1, 1, 0], #J
    [1, 1, 1, 1, 0, 1, 1, 1, 1]  #O
]
y = [
    [1, 0, 0],
    [0, 1, 0],
    [0, 0, 1]
]
learning_rate = 0.01
```

```python
ill = len(X[0])
oll = len(y[0])
```

```python
weights = [ [random.uniform(-1, 1) for i in range(ill)] for j in range(oll) ]
biases = [ random.uniform(-1, 1) for j in range(oll) ]
print(f'weights: {weights}\nbiases: {biases}')
```

```python
def predict(I):
    z = [ 
        sum([ I[i] * weights[j][i] for i in range(ill) ]) + biases[j] for j in range(oll)
    ]
    output = [ sigmoid(z[j]) for j in range(oll) ]
    return z, output
```

```python
costHistory = []

cost = 1
while cost > 0.001:
    costs = []
    for p in range(X.__len__()):
        I = X[p]
        E = y[p]

        z, output = predict(I)
        # z = [ 
        #     sum([ I[i] * weights[j][i] for i in range(ill) ]) + biases[j] for j in range(oll)
        # ]
        # output = [ sigmoid(z[j]) for j in range(oll) ]

        Cw = [ [ 2 * (output[j]-E[j]) * dsigmoid(z[j]) * I[i] for i in range(ill) ] for j in range(oll) ]
        Cb = [ 2 * (output[j]-E[j]) * dsigmoid(z[j]) for j in range(oll) ]
        
        weights = [ [ weights[j][i] - Cw[j][i] * learning_rate for i in range(ill) ] for j in range(oll) ]
        biases = [ biases[j] - Cb[j] * learning_rate for j in range(oll) ]

        costs.append(sum([ math.pow(output[j]-E[j], 2) for j in range(oll) ]))

    cost = sum(costs)/len(costs)
    costHistory.append(cost)

print(f'Cost: {cost}')

# plt.subplot(1, 2, 1)
# plt.plot([i for i in range(len(costHistory))], costHistory)
# plt.title('Normal Scale')
# plt.xlabel('Number of Iterations')
# plt.ylabel('cost')

# plt.subplot(1, 2, 2)
# plt.loglog([i for i in range(len(costHistory))], costHistory)
# plt.title('Logarithmic Scale')
# plt.xlabel('Number of Iterations')
# plt.ylabel('cost')

# plt.show()
```
